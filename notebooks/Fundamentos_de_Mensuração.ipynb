# Mensuração

Mensuração de dados é um conjunto de ações realizadas pelas empresas para compreender o desempenho das estratégias praticadas por elas. É uma forma de saber se correspondem com os objetivos traçados. É uma prática fundamental para a gestão de negócios.


# Sua importância

* Permite saber se os objetivos traçados estão sendo alcançados
* Identifica ações que trazem os melhores resultados
* Permite otimizar recursos e esforços da equipe
* Permite identificar áreas de sucesso e oportunidades de melhoria
* Permite tomar decisões embasadas para otimizar o desempenho futuro


# Tipos de escala de mensuração

Os tipos de escala são: nominal, ordinal, intervalo e razão. São técnicas que ajudam a padronizar e agilizar a coleta de dados.


Escala Nominal
* Identifica e Classifica objetos
* Exemplos: sexo, raça, tipo sanguíneo

Escala Ordinal
* Indica as posições relativas dos objetos
* Exemplos: tamanho (pequeno, médio, grande), classe social (baixa, média, alta)

Escala Intervalar
* Escala numérica que não apresenta zero absoluto
* Exemplos: temperatura, horário do dia

Escala de Razão
* Escala numérica que apresenta zero absoluto, significando a “falta de”
* Exemplos: renda familiar, preço de uma casa




# Exemplo Prático

Nesse exemplo vamos explorar as: medidas de tendência central (média, mediana e moda), medidas de dispersão(variância, desvio padrão, amplitude e coeficiente de variação), medidas de posição(quartis, decis e percentis), medidas de associação(Covariância, Correlação (Pearson, Spearman, Kendall)) e visualização e interpretação(histogramas, boxplots, scatter plots para correlação).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Categoria': ['A', 'B', 'C', 'D', 'E'] * 20,
    'Valor': np.random.randint(10, 100, 100)
}
df = pd.DataFrame(data)
```

```python
media = df['Valor'].mean()
mediana = df['Valor'].median()
moda = df['Valor'].mode()[0]
```

```python
variancia = df['Valor'].var()
desvio_padrao = df['Valor'].std()
amplitude = df['Valor'].max() - df['Valor'].min()
coeficiente_variacao = desvio_padrao / media
```

```python
quartis = df['Valor'].quantile([0.25, 0.5, 0.75])
```

```python
plt.figure(figsize=(12,5))
sns.histplot(df['Valor'], bins=10, kde=True)
plt.title('Distribuição dos Valores')
plt.show()

sns.boxplot(x=df['Valor'])
plt.title('Boxplot dos Valores')
plt.show()
```

```python
print(f"Média: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
print(f"Variância: {variancia}")
print(f"Desvio Padrão: {desvio_padrao}")
print(f"Amplitude: {amplitude}")
print(f"Coeficiente de Variação: {coeficiente_variacao}")
print(f"Quartis:\n{quartis}")
```

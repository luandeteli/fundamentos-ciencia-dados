# Análise de Correlação

A correlação é uma medida estatística que indica a relação entre duas variáveis. Ela pode ser positiva, negativa ou inexistente.


Muito se confunde os termos correlação e causalidade, mas são termos muito diferentes. Correlação é uma relação entre variáveis, enquanto causalidade é uma relação entre causa e efeito entre variáveis.

# Principais caracteristicas de correlação e causalidade:

Correlação

* Descreve uma associação entre variáveis, ou seja, quando uma variável muda, a outra também muda

* É um indicador estatístico da relação entre variáveis

* Não implica causalidade, ou seja, correlação não significa que a mudança em uma variável é a causa da mudança na outr


Causalidade

* Indica que um evento é o resultado da ocorrência do outro evento

* É uma declaração mais forte do que correlação

* Para estabelecer causalidade, é necessário um experimento controlado


Um exemplo de correlação que não implica causalidade é a relação entre vendas de sorvete e afogamentos em piscinas. Ambas as variáveis aumentam durante o verão, mas isso não significa que comer sorvete causa afogamentos. A relação pode ser explicada por temperaturas mais quentes.


Confundir correlação com causalidade é um erro comum e pode levar à falácia da causa falsa.



Os três tipos de correlação que existe é: Positiva, Negativa e Inexistente.


Correlação positiva

* Quando uma variável aumenta, a outra também tende a aumentar

* O coeficiente de correlação (r) é maior que 0, até 1

* Exemplo: renda e consumo


Correlação negativa

* Quando uma variável aumenta, a outra tende a diminuir

* O coeficiente de correlação (r) é menor que 0, até -1

* Exemplo: velocidades maiores no carro e tempo de viagem menor


Correlação inexistente

* Não há uma relação linear evidente entre as variáveis

* O coeficiente de correlação (r) é aproximadamente 0

* Exemplo: ganho de peso e pontuações nos exames escolares

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
```

```python
np.random.seed(42)
data = {
    'Idade': np.random.randint(18, 60, 100),
    'Salário': np.random.randint(2000, 10000, 100),
    'Gastos': np.random.randint(1000, 8000, 100)
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
correlation_matrix = df.corr()
```

```python
print("Matriz de Correlação:\n", correlation_matrix)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlação")
plt.show()
```

```python
sns.lmplot(x='Salário', y='Gastos', data=df)
plt.title("Relação entre Salário e Gastos")
plt.show()
```

```python
sns.pairplot(df, kind="reg")
plt.suptitle("Pairplot das Variáveis", y=1.02)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.regplot(x='Idade', y='Salário', data=df, ax=axes[0])
axes[0].set_title("Correlação entre Idade e Salário")

sns.regplot(x='Idade', y='Gastos', data=df, ax=axes[1])
axes[1].set_title("Correlação entre Idade e Gastos")

sns.regplot(x='Salário', y='Gastos', data=df, ax=axes[2])
axes[2].set_title("Correlação entre Salário e Gastos")

plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['Idade'], df['Salário'], df['Gastos'], c=df['Gastos'], cmap='viridis', marker='o')

ax.set_xlabel("Idade")
ax.set_ylabel("Salário")
ax.set_zlabel("Gastos")
ax.set_title("Gráfico de Dispersão 3D")

plt
```

# O que é o BI (Business Inteligence)?

BI, ou Business Inteligence, é o processo de coletar e analisar dados para que as organizações possam tomar decisões estratégicas e diárias. Refere-se ao uso de dados para apoiar a tomada de decisão nas empresas. Utilizando técnicas de visualização de dados, é possível extrair insights valiosos. Também é conhecido como sistema de suporte de descisão (DSS)

O BI pode ajudar as empresas a: Tomar decisões baseadas em dados, Eliminar ineficiências, Adaptar-se a mudanças no mercado, Implementar processos de negócios mais eficientes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
sns.set_theme(style="darkgrid")
```

```python
df = pd.DataFrame({
    'Mês': ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun'],
    'Receita': [15000, 18000, 12000, 22000, 24000, 19000],
    'Custo': [7000, 8000, 6000, 11000, 13000, 9000]
})

df['Lucro'] = df['Receita'] - df['Custo']
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Mês', y='Receita', data=df, palette='Blues')
plt.title('Receita por Mês')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(df['Custo'], labels=df['Mês'], autopct='%1.1f%%', colors=sns.color_palette('coolwarm'))
plt.title('Distribuição de Custos')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Receita', 'Custo', 'Lucro']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlação entre Variáveis')
plt.show()
```

```python
fig = px.scatter(df, x='Receita', y='Lucro', text='Mês', size='Lucro', color='Mês')
fig.update_traces(textposition='top center')
fig.show()
```

# Conclusão

Com essas visualizações, podemos entender melhor os dados e tomar decisões mais informadas. A combinação de BI com técnicas de visualização torna a análise mais intuitiva e eficaz.

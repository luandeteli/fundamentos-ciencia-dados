# Estatística e Probabilidade

Estatística é uma área da matemática que se dedica à coleta, organização, análise e interpretação de dados. Ela nos permite entender, resumir e tirar conclusões a partir de grandes volumes de informações. A estatística é amplamente utilizada para identificar padrões, fazer previsões e tomar decisões baseadas em dados. Ela pode ser dividida em duas grandes áreas: descritiva e inferencial. A estatística descritiva se ocupa da organização e resumo dos dados, enquanto a inferencial aplica métodos para tirar conclusões sobre uma população com base em uma amostra.

Já a Probabilidade é a área da matemática que estuda os fenômenos aleatórios. Ela nos ajuda a calcular a chance de um evento ocorrer. A probabilidade é fundamental para a tomada de decisões em situações incertas, permitindo que possamos quantificar e gerenciar riscos. Ela fornece as bases para muitas distribuições de probabilidade, como a distribuição normal, binomial e outras, que são usadas na modelagem e previsão de diversos fenômenos.


# Sua Importância

Estatística e probabilidade são fundamentais para a compreensão e análise de dados no mundo atual. Com o aumento do volume de informações geradas diariamente, essas áreas se tornam essenciais em praticamente todos os campos do conhecimento e da indústria, como:

* Pesquisa científica: Estatísticas são utilizadas para analisar experimentos e tirar conclusões válidas

* Negócios e Marketing: A análise de dados permite entender o comportamento do consumidor, realizar previsões de vendas e otimizar campanhas publicitárias

* Medicina: Probabilidade e estatística são usadas em estudos clínicos, diagnóstico de doenças e avaliação de tratamentos.

* Engenharia: Em áreas como controle de qualidade e design de experimentos, a estatística ajuda a garantir que os produtos atendam aos padrões exigidos.

Além disso, a compreensão dos conceitos de probabilidade e estatística é essencial para tomar decisões informadas em um mundo cada vez mais orientado por dados.

# Exemplo Prático

O objetivo deste projeto é explorar e aplicar os conceitos fundamentais de Estatística e Probabilidade através de uma análise prática de dados. Utilizando distribuições de probabilidade, como a distribuição normal e binomial, buscamos compreender os comportamentos de dados aleatórios e como esses fenômenos podem ser modelados e analisados matematicamente.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
mean = 0
std_dev = 1
size = 1000
data_normal = np.random.normal(loc=mean, scale=std_dev, size=size)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data_normal, kde=True)
plt.title("Distribuição Normal")
plt.show()
```

```python
mean_data = np.mean(data_normal)
median_data = np.median(data_normal)
std_data = np.std(data_normal)
mean_data, median_data, std_data
```

```python
n = 10
p = 0.5
size = 1000
data_binomial = np.random.binomial(n=n, p=p, size=size)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data_binomial, kde=False, bins=20)
plt.title("Distribuição Binomial")
plt.show()
```

```python
group1 = np.random.normal(loc=5, scale=1, size=100)
group2 = np.random.normal(loc=5.5, scale=1, size=100)
```

```python
t_stat, p_value = stats.ttest_ind(group1, group2)

t_stat, p_value
```

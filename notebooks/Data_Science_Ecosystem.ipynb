# O Ecossistema de Data Science

O ecossistema de ciência de dados é o conjunto de tecnologias, ferramentas, processos e pessoas que permitem a coleta, armazenamento, análise e visualização de dados. É uma rede dinâmica e multifacetada que envolve uma combinação de ferramentas, técnicas e habilidades necessárias para coletar, processar, analisar e interpretar dados de forma eficaz. Com o aumento da quantidade de dados disponíveis e a crescente demanda por insights acionáveis, entender como funciona esse ecossistema é fundamental para qualquer profissional da área. Data Science não se resume apenas ao uso de algoritmos complexos; ele engloba um conjunto de processos que incluem a preparação dos dados, a escolha das ferramentas apropriadas, a aplicação de modelos estatísticos e a apresentação de resultados.

# Principais Componentes

* Linguagens de Programação: Python, R, SQL e Julia
* Bibliotecas e frameworks: Pandas, NumPy, Matplotlib, Seaborn, Scikit-learn, TensorFlow, PyTorch, entre outras
* Ferramentas e Plataformas: Jupyter Notebook, Google Colab, Apache Spark, Docker, GitHub
* Fluxo de Trabalho: Coleta, limpeza, exploração, modelagem e visualização dos dados

# Exemplo Prático

Vamos utilizar algumas dessas ferramentas do ecossistema de Data Science para uma análise simples dos dados. A ideia é sempre praticar para ter capacidade de abstração.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Nome': ['Alice', 'Bob', 'Carlos', 'Diana', 'Eduardo'],
    'Idade': [25, 30, 35, 40, 22],
    'Salário': [5000, 7000, 8000, 6500, 4000],
    'Área': ['TI', 'Marketing', 'Financeiro', 'RH', 'TI']
}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Salário'], bins=5, kde=True, color='blue')
plt.title("Distribuição de Salários")
plt.xlabel("Salário")
plt.ylabel("Frequência")
plt.show()
```

```python
df_grouped = df.groupby('Área')['Salário'].mean().reset_index()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df_grouped, x='Área', y='Salário', palette='viridis')
plt.title("Média Salarial por Área")
plt.xlabel("Área")
plt.ylabel("Salário Médio")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Idade', y='Salário', hue='Área', palette='coolwarm', s=100)
plt.title("Relação entre Idade e Salário")
plt.xlabel("Idade")
plt.ylabel("Salário")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Área', palette='pastel')
plt.title("Quantidade de Profissionais por Área")
plt.xlabel("Área")
plt.ylabel("Quantidade")
plt.show()
```

```python
df_std = df.groupby('Área')['Salário'].std().reset_index()
plt.figure(figsize=(8, 5))
sns.barplot(data=df_std, x='Área', y='Salário', palette='muted')
plt.title("Desvio Padrão dos Salários por Área")
plt.xlabel("Área")
plt.ylabel("Desvio Padrão do Salário")
plt.show()
```
